```python
import networkx as nx
import plotly.graph_objects as go

def visualize_graph(data: list[dict]):
# Data provided by user

    # Define groups for coloring
    id_to_group = {
        'Lisa': 'Person', 'Emily': 'Person',
        'Picnic': 'Event', 'Outing': 'Event', 'Past Picnic Mishap': 'Event',
        'Blanket': 'Object',
        'Sandwiches': 'Food', 'Strawberries': 'Food', 'Lemon Drizzle Cake': 'Food', 'Bread': 'Food',
        'Homemade Lemonade': 'Drink',
        'Content': 'Concept', 'Grateful': 'Concept',
        'Squirrel': 'Animal',
        'Elmwood Park': 'Location',
        'Afternoon': 'Time',
        'Oak Tree': 'Plant'
    }

    # Create NetworkX graph
    G = nx.DiGraph()

    for row in data:
        n = row['n']
        m = row['m']
        r = row['r']
        
        n_id = n['id']
        m_id = m['id']
        r_type = r[1]
        
        n_group = id_to_group.get(n_id, 'Unknown')
        m_group = id_to_group.get(m_id, 'Unknown')
        
        G.add_node(n_id, label=n_id, group=n_group)
        G.add_node(m_id, label=m_id, group=m_group)
        G.add_edge(n_id, m_id, label=r_type)

    # Generate positions
    pos = nx.spring_layout(G, seed=42, k=0.5) # k adjusts spacing

    # Create Plotly traces
    edge_x = []
    edge_y = []
    edge_text = []

    for edge in G.edges(data=True):
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.append(x0)
        edge_x.append(x1)
        edge_x.append(None)
        edge_y.append(y0)
        edge_y.append(y1)
        edge_y.append(None)
        edge_text.append(edge[2]['label'])

    edge_trace = go.Scatter(
        x=edge_x, y=edge_y,
        line=dict(width=0.5, color='#888'),
        hoverinfo='text',
        text=edge_text,
        mode='lines')

    node_x = []
    node_y = []
    node_display_text = []
    node_hover_text = []
    node_color = []

    # Map groups to colors
    groups = list(set(nx.get_node_attributes(G, 'group').values()))
    group_to_color = {g: i for i, g in enumerate(groups)}

    for node in G.nodes(data=True):
        node_id = node[0]
        node_attrs = node[1]
        x, y = pos[node_id]
        node_x.append(x)
        node_y.append(y)
        
        # Display text (on graph)
        node_display_text.append(node_attrs['label'])
        
        # Hover text (tooltip) - Include relationships
        connections = []
        # Outgoing
        for neighbor in G.successors(node_id):
            edge_data = G.get_edge_data(node_id, neighbor)
            connections.append(f"➡ {edge_data['label']} ➡ {neighbor}")
        # Incoming
        for neighbor in G.predecessors(node_id):
            edge_data = G.get_edge_data(neighbor, node_id)
            connections.append(f"⬅ {edge_data['label']} ⬅ {neighbor}")
            
        hover_str = f"<b>{node_attrs['label']}</b> ({node_attrs['group']})<br>" + "<br>".join(connections)
        node_hover_text.append(hover_str)
        
        node_color.append(group_to_color.get(node_attrs['group'], 0))

    node_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='markers+text',
        text=node_display_text,
        hovertext=node_hover_text,
        textposition="bottom center",
        hoverinfo='text',
        marker=dict(
            showscale=True,
            colorscale='YlGnBu',
            color=node_color,
            size=20,
            colorbar=dict(
                thickness=15,
                title='Node Group',
                xanchor='left',
            ),
            line_width=2))

    fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='Neo4j Graph Visualization',
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=20,l=5,r=5,t=40),
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                    )
    fig.show()
```
